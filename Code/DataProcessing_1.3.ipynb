```python
# Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../Source Data/data.csv')
df.head()
```

```python
# Shape of dataframe
# print(" Shape of dataframe: ", df.shape)
```

```python
# Drop duplicates
df = df.drop_duplicates()
# print(" Shape of dataframe after dropping duplicates: ", df.shape)
```

```python
# Variable inspection

# print("Names of columns ", list(df.columns))
```

```python
# Drop unnecessary columns
df = df.drop(columns= ["Unnamed: 0", "ID", "Name", "Photo", "Nationality", "Flag", "Club", "Club Logo", "Value", "Wage", "Potential", "Special", "Preferred Foot", "International Reputation", "Skill Moves", "Work Rate", "Body Type", "Real Face", "Jersey Number", "Joined", "Loaned From", "Contract Valid Until", "Height", "Weight", "LS", "ST", "RS", "LW", "LF", "CF", "RF", "RW", "LAM", "CAM", "RAM", "LM", "LCM", "CM", "RCM", "RM", "LWB", "LDM", "CDM", "RDM", "RWB", "LB", "LCB", "CB", "RCB", "RB", "Release Clause"])
df.head()
```

```python
print("Names of columns ", list(df.columns))
```

```python
#Null values

null= df.isnull().sum().sort_values(ascending=False)
total =df.shape[0]
percent_missing= (df.isnull().sum()/total).sort_values(ascending=False)

missing_data= pd.concat([null, percent_missing], axis=1, keys=['Total missing', 'Percent missing'])

missing_data.reset_index(inplace=True)
missing_data= missing_data.rename(columns= { "index": " column name"})
 
print ("Null Values in each column:\n", missing_data)
```

```python
df_isnull = pd.isnull(df["Position"])
# df[df_isnull]
```

```python
#Filtering data with null values for position

# df_isnull = pd.isnull(df["LB"])
pos_null= df[df_isnull]
print(pos_null.shape)
print(pos_null.isnull().sum().sort_values(ascending= False))
```

```python
#Filtering relevant data & checking for club null values
df_notnull = pd.notnull(df["Position"])
df = df[df_notnull]
print(df.shape)
# df.head()
```

```python
# Partitioning original player data according to positions
# We mainly divide players into 4 positions: Forward, Midfielder, Back and Goalkeeper
# print(df["Position"].unique())
df_fwd = df.loc[df["Position"].isin(["CF", "LF", "RF", "ST", "LS", "RS"])].drop(columns=['Position'])
df_md = df.loc[df["Position"].isin(["LM", "RM", "CM", "LCM", "RCM", "CAM", "LAM", "RAM", "CDM", "LDM", "RDM", "LW", "RW"])].drop(columns=['Position'])
df_bk = df.loc[df["Position"].isin(["CB", "LB", "RB", "LCB", "RCB", "LWB", "RWB"])].drop(columns=['Position'])
df_gk = df.loc[df["Position"].isin(["GK"])].drop(columns=['Position'])
df_fwd.head()
```

```python
df.drop(columns=['Position']).to_csv("../Processed Data/AllPlayerDataProcessed.csv", index_label = "Index")
df_fwd.to_csv("../Processed Data/FWDDataProcessed.csv", index_label = "Index")
df_md.to_csv("../Processed Data/MDDataProcessed.csv", index_label = "Index")
df_bk.to_csv("../Processed Data/BKDataProcessed.csv", index_label = "Index")
df_gk.to_csv("../Processed Data/GKDataProcessed.csv", index_label = "Index")
```
